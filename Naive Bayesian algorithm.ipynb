```python
import pandas as pd
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import LabelEncoder
```

```python
data = {
    'Age' : ['youth', 'youth', 'middle', 'senior', 'senior', 'middle', 'senior','youth', 'youth', 'senior', 'youth', 'middle', 'middle', 'senior'],
    'Income' : ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high','medium'],
    'Student' : ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'no', 'yes','no'],
    'CreditRating' : ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair','fair', 'fair', 'excellent', 'excellent', 'fair','excellent'],
    'BuysComputer' : ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'no','yes']
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
le = {}
df_enc = df.copy()
for col in df.columns:
  le[col] = LabelEncoder()
  df_enc[col] = le[col].fit_transform(df[col])
```

```python
df_enc
```

```python
x = df_enc[['Age','Income','Student','CreditRating']]
y = df_enc['BuysComputer']
```

```python
x
```

```python
y
```

```python
model = CategoricalNB()
algo=model.fit(x,y)
```

```python
test =[['Youth','Medium','Yes','Fair']]
```

```python
test
```

```python
print("\n--------Prediction---------------------------")
for t in test:
  row = [[
      le['Age'].transform([t[0].lower()])[0],
      le['Income'].transform([t[1].lower()])[0],
      le['Student'].transform([t[2].lower()])[0],
      le['CreditRating'].transform([t[3].lower()])[0]
  ]]

  pred = le['BuysComputer'].inverse_transform(algo.predict(row))[0]
  prob = model.predict_proba(row)[0]
  print(f"Input: Age={t[0]}, Income={t[1]}, Student={t[2]}, CreditRating={t[3]}")
  print(f"Result: BuysComputer={pred}")
  print(f"P(No) = {prob[0]:.4f}")
  print(f"P(Yes) = {prob[1]:.4f}")
```

```python
from sklearn.metrics import accuracy_score
y_pred = model.predict(x)
print(f"\nTraining Accuracy: {accuracy_score(y,y_pred)*100:.2f}%")
```
